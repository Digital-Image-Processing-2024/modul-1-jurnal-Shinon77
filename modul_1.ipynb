## Normalisasi Citra atau Gambar

```python
norm_image = gray_image

jum_kemunculan = {}
for i in range(256):
    jum_kemunculan[i] = 0

for i in range(norm_image.shape[0]):
    for j in range(norm_image.shape[1]):
        jum_kemunculan[norm_image[i, j]] += 1

prob = {}
print('Probabilitas:')
for i in range(256):
    prob[i] = jum_kemunculan[i] / gray_image.size
    print(f"{i}: {prob[i]}")
```

Code diatas bertujuan untuk menormalisasikan nilai setiap index yang ada pada citra grayScale. Normalisasi disini dilakukan dengan cara membuat sebuah matriks yang nantinya diisi oleh nilai dari masing-masing index `gray_image`. Cara menormalisasi disini dapat dilakukan dengan cara mencari probability (hi). Rumus hi yaitu jumlah kemunculan nilai (ni) dibagi dengan jumlah seluruh index yang ada (n).

## Konvolusi dan Tresholding

```python
kernelMatrix = np.array([[1, 2,-1, 1],
                         [0, 2, 1, 2],
                         [2, 1, 1, 0],
                         [1, 0, -1, 0]])
```

Code diatas membuat sebuah `kernelMatrix`. Kernel ini dibuat dengan menggunakan `np.array()` yang dimana akan berbentuk matriks 4 x 4. Library yang digunakan yaitu numpy.

```python
padded_img = np.zeros((gray_image.shape[0] + 2, gray_image.shape[1] + 2))
padded_img[1:-1, 1:-1] = gray_image
print(padded_img)
```

Code diatas berfungsi untuk memberikan padding zero pada semua sisi matriks sehingga kernel yang berlebih dapat terisi dengan menggunakan padding. Padding ini bertujuan untuk mempermudah operasi Sobel bekerja dengan benar pada tepi citra.

## Menampilkan Hasil Konvolusi dan Tresholding

```python
imgRow, imgCol = gray_image.shape
kernelRow, kernelCol= kernelMatrix.shape
```

Code diatas melakukan inisialisasi ukuran tinggi dan lebar dari sebuah citra dengan menggunakan fungsi `.shape`. Kemudian kode diatas juga melakukan inisialisasi dimensi baris dan kolom dari matriks kernel. Dimensi ini digunakan untuk menentukan berapa banyak iterasi yang diperlukan saat melakukan operasi Sobel pada citra.

```python
pad = np.zeros((imgRow - kernelRow + 1, imgCol - kernelCol + 1))
```

Code diatas membuat sebuah array pad dengan dimensi yang lebih kecil dari citra asli. Dimensi pad ini dihitung dengan menggunakan dimensi kernel dari dimensi citra dan menambahkannya dengan 1.

```python
for i in range(pad.shape[0]):
    for j in range(pad.shape[1]):
        pad[i][j] = round((gray_image[i:i + kernelRow, j:j + kernelCol] * kernelMatrix).sum())
```

Code diatas merupakan sebuah perulangan yang dimana terdapat proses perkalian perelemen antara matriks citra dengan kernel. Kemudian hasil dari perkalian tersebut akan dijumlahkan dan dibulatkan untuk menghasilkan gradien pada piksel yang sedang diproses.

Code dibawah berfungsi untuk menampilkan hasil konvolusi. Dari hasil perhitungan operasi Sobel sebelumnya akan dibagi dengan nilai maximal dan dikurangi dengan nilai minimal yang kemudian membentuk sebuah citra seperti dibawah ini.

```python
min = np.min(pad)
max = np.max(pad)
convolution = np.round((pad - min)/max-min)

plt.subplot(1,2,1)
plt.title('Before Convolution')
plt.imshow(pad, cmap='gray')

plt.subplot(1,2,2)
plt.title('Convolution Result')
plt.imshow(convolution, cmap='gray')
```

Code dibawah berfungsi untuk melakukan rotasi citra dengan derajat rotasinya yaitu 320 derajat (searah jarum jam). Kemudian menggunakan rumus yang telah diinisialisasi, `matrix_trans` akan dihitung berdasarkan sudut yang diberikan. Setelah itu piksel baru untuk citra dirotasi dihitung dan diisi dengan memastikan titik koordinat berada dalam dimensi citra. Kemudian hasil rotasi akan ditampilkan menggunakan `plt.imshow(rotation_image, cmap='gray')` dengan judul `Rotation Image`.

```python
rotation_corner = 320

sin_theta = np.sin(np.radians(rotation_corner))
cos_theta = np.cos(np.radians(rotation_corner))

matrix_trans = np.array([[cos_theta, -sin_theta],
                         [sin_theta, cos_theta]])

rotation_image = np.zeros_like(gray_image)
for i in range(gray_image.shape[0]):
    for j in range(gray_image.shape[1]):
        newX = int(round(cos_theta * (j - gray_image.shape[1] // 2) - sin_theta * (i - gray_image.shape[0] // 2) + gray_image.shape[1] // 2))
        newY = int(round(sin_theta * (j - gray_image.shape[1] // 2) + cos_theta * (i - gray_image.shape[0] // 2) + gray_image.shape[0] // 2))

        if 0 <= newX < gray_image.shape[1] and 0 <= newY < gray_image.shape[0]:
            rotation_image[i, j] = gray_image[newY, newX]

plt.title('Rotation Image')
plt.imshow(rotation_image, cmap='gray')
```

Setelah dilakukannya rotasi, sekarang untuk code dibawah merupakan proses translasi. Pada proses ini, citra yang telah dirotasi akan digeser dengan nilai spesifik yang telah diberikan. ## Mirorring dan Dilatasi

Code dibawah ini merupakan sebuah proses pencerminan atau mirorring. Gambar yang telah di translasi sebelumnya akan dibalik dengan irisan `[:, ::, -1]`. Irisan ini digunakan untuk membalik urutan kolom citra dan menghasilkan efek cermin pada citra tersebut.

```python
image_flip = img_translasi[:, ::-1]
plt.imshow(image_flip, cmap="gray")
```

Code dibawah ini merupakan sebuah proses dilatasi pada citra yang telah dimirorring. Array akan dibuat dengan dimensi yang diperbesar berdasarkan skala yang telah diberikan. Setiap piksel pada citra asli digandakan atau direplikasi skala kali di kedua arah. Hal ini menghasilkan gambar yang diperbesar dengan detail yang lebih jelas.

```python
scale = 3

image_dilatasi = np.zeros((image_flip.shape[0] * scale, image_flip.shape[1] * scale), dtype=image_flip.dtype)
for i in range(image_flip.shape[0]):
    for j in range(image_flip.shape[1]):
        for k in range(scale):
            for l in range(scale):
                image_dilatasi[i * scale + k, j * scale + l] = image_flip[i, j]

plt.title('Gambar Didilatasi (Skala ' + str(scale) + '):')
plt.imshow(image_dilatasi, cmap='gray')
```
